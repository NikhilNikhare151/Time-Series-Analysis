# Time Series Analysis - Electricity Price & Demand using ARIMA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.arima.model import ARIMA

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"E:\Artificial Intelligence\dataset\Electricity Price & Demand.csv")
df.head()
```

```python
demand = df["TOTALDEMAND"]
demand
```

```python
plt.figure(figsize=(15,4))
plt.plot(demand)
plt.title("Demand Trend")
plt.grid()
plt.show()
```

```python
from pmdarima import auto_arima
```

```python
model = auto_arima(demand, seasonal=False, stepwise=True, suppress_warnings=True)
```

```python
model.order
```

```python
model = ARIMA(demand, order=model.order)
```

```python
result = model.fit()
```

```python
forecast = result.get_forecast(steps=500)
```

```python
plt.figure(figsize=(15,4))
plt.plot(demand)
plt.plot(forecast.predicted_mean)
plt.title("Demand Trend")
plt.grid()
plt.show()
```
